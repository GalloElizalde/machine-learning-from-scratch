# Autoencoder for credit card detection (Pytorch)

The purpose of this notebook is to develop a model capable of identifying fraudulent credit card transactions within a dataset largely composed of legitimate operations. To achieve this, we use an autoencoder, a specific type of neural network that has the same number of input and output nodes and several lower-dimensional hidden layers that impose a bottleneck representation.

The dataset used, available on [Kaggle](https://www.kaggle.com/mlg-ulb/creditcardfraud) (big file: 144 MB), contains credit card transactions recorded over a two-day period, totaling 284,807 operations, of which 492 were labeled as fraud.

All 30 features in the dataset are numerical. For confidentiality reasons, most variables were transformed using PCA, with the exceptions of Time and Amount, which remain unaltered. Time indicates the number of seconds elapsed between each transaction and the first entry in the dataset.

The dataset also contains the class of event: 0 = normal transaction; 1 = fraudulous transaction.
