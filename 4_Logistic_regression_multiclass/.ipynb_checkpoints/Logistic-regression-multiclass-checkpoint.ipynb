# Multiclass classification with the Iris dataset

The Iris dataset consists of 3 different classes of irises (Setosa, Versicolour, and Virginica). 

Each iris is described by four features: Sepal Length, Sepal Width, Petal Length and Petal Width. 

The objective of this exercice is to train a multiclass classification model using the [logistic regression](https://scikit-learn.org/stable/modules/generated/sklearn.linear_model.LogisticRegression.html) model available in `scikit-learn`.

## Initialisation

Load the following libraries.

````python
import numpy as np
import matplotlib
import matplotlib.pyplot as plt
%matplotlib inline

from sklearn import datasets
from sklearn.model_selection import train_test_split

from sklearn.linear_model import LogisticRegression
from sklearn.metrics import accuracy_score
````

## Data exploration

The dataset contains 150 examples, stored in 150x4 numpy array. The rows being the samples and the columns being the features. As there are 3 classes of irises, each example corresponds to a target value: `0 = setosa, 1 = versicolor, 2 = virginica`.

````python
# Load Iris dataset

iris = datasets.load_iris()


# separate data into target and feature

X = iris.data   # features
t = iris.target # target


# target and feature names

feature_names = iris.feature_names
species = iris.target_names

print("Iris species:", species)
print("Features name:",feature_names)
````

````python
# Make a histogram of each feature of the data

plt.figure(figsize=(18,11))
colors = ["purple", "green", "yellow"]
for i in range(4): # cicle over each feature i
    plt.subplot(2,2,i+1)
    plt.title(feature_names[i])
    for j in range(3): # cicle over each class
        plt.hist(X[:,i][t == j], label = species[j] , color = colors[j], alpha = 0.45, bins = 15)
    plt.grid(True, axis = "y")
    plt.xlabel("cm")
    plt.ylabel("Frequency")
    plt.legend()

plt.savefig("distribution_feautures.png", dpi = 300)
````

````python
# Show the scatter plots of each feature versus a different feature

plt.figure(figsize=(17,11))
cnt = 1
for i in range(4):
    for j in range(i+1,4):
        plt.subplot(2,3,cnt)
        plt.grid(True)   
        scatter = plt.scatter(X[:,i], X[:,j], c = t, edgecolor = "k")
        plt.xlabel(feature_names[i])
        plt.ylabel(feature_names[j])
        cnt = cnt + 1
        plt.legend(*scatter.legend_elements())
         
plt.savefig("scatter_plot_feature_vs_feature.png", dpi = 300)
````

## Train model and predict

Logistic regression model described here: [LogisticRegression](https://scikit-learn.org/stable/modules/generated/sklearn.linear_model.LogisticRegression.html)


````python
# Split the dataset: 2 third sample size, 1 third for testing

data_train, data_test, labels_train, labels_test = train_test_split(X, t, test_size = .33, random_state = 42, shuffle = True) # Use the train_test_split method
print(data_test.shape)


# Initialize the logistic regression model

model = LogisticRegression(solver="lbfgs", multi_class='multinomial', penalty='l2',max_iter=500)


# Train the model on the training set

model.fit(data_train, labels_train)


# Display the values of the weights and the bias terms

print("Weights:", model.coef_) # Coefficients (weights for each feature and class)
print("\nbiases terms:", model.intercept_)  # # Intercepts (bias terms for each class)


# Predict the class label for all samples in the training and test datasets

labels_train_prediction = model.predict(data_train)
print('\n The accuracy of Logistic Regression is (train dataset): %.3f' % accuracy_score(labels_train_prediction, labels_train)) # accuracy of prediction
labels_test_prediction = model.predict(data_test)
print('\n The accuracy of Logistic Regression is (train dataset): %.3f' % accuracy_score(labels_test_prediction, labels_test))  # accuracy of prediction


# For each class show a histogram of the probability estimate of being in the right class. 

probability_test = model.predict_proba(data_test)
plt.hist(probability_test[labels_test == 0][:,0], alpha = .45, label = feature_names[0], bins = "auto")
plt.hist(probability_test[labels_test == 1][:,1], alpha = .45, label = feature_names[1], bins = "auto")
plt.hist(probability_test[labels_test == 2][:,2], alpha = .45, label = feature_names[2], bins = "auto")
plt.xlabel("Probability")
plt.ylabel("Frequency")
plt.title("Probability score for each class")
plt.grid(True, axis = "y")
plt.legend()

plt.savefig("probability_score_per_class_test.png", dpi = 300)
````

## How probability is calculated

In this part we redo the calculation of the probability on the test dataset by using **only** `numpy` tools.

Fisrt we explain how the predicted class (the calculated class label) is determined for each example:



1. Probability estimate
For each of the $k={1..K}$ classes calculate:
$s_k = w_k^T.x + b_k$

    Probability estimate for each class:

$$\text{probability}_k = \text{softmax}(s_k) = \frac{exp(s_k)}{\sum_{j=1}^K exp(s_j)}$$

2. Predicted label

    In principle given by the position of the highest value in the probability estimate. However in scikit-learn the predict function calculates the label based on the signed distance of the sample to the decision hyperplane, that is just $s_k = w_k^T.x + b_k$:
```
        scores = safe_sparse_dot(X, self.coef_.T, dense_output=True) + self.intercept_
        return scores.ravel() if scores.shape[1] == 1 else scores
        indices = scores.argmax(axis=1)
```

````python
# Functions

def sigmoid(x):
    return 1. / (1. + np.exp(-x))

def softmax(x):
    return (np.exp(x))/(np.exp(x).sum())

def probability_numpy(x,weights,bias):
    xlin = np.matmul(weights,x) + bias
    return softmax(xlin)

def predicted_class(probability):
    return probability.argmax()

````

````python
# Calculate probability for each class using numpy

score = probability_numpy(data_test[5],model.coef_,model.intercept_)
#index = score.argmax(axis=1)
print(score)
print(score.argmax())
````

````python
# Example for few test sample

for i in np.arange(0,10,1):
    x = data_test[i]
    prob_numpy = probability_numpy(x,model.coef_,model.intercept_)
    prob_logreg = probability_test[i]
    print('Example %d: y_test = %d' %(i,labels_test[i]))
    print("  probability estimate (log. regr.):",prob_logreg)
    # for multiclass softmax is used in place of sigmoid to calculate probabilities
    print("  probability estimate (using numpy):",prob_numpy," --> class = %d" %predicted_class(prob_numpy))
````

## Plot decision surfaces

Show the decision surfaces on the scatter plots of each feature versus another, with different color for each classes.

````python
# Show decision boundaries:
# https://scikit-learn.org/dev/auto_examples/linear_model/plot_logistic_multinomial.html#sphx-glr-auto-examples-linear-model-plot-logistic-multinomial-py

c=['tab:blue','tab:orange','tab:green']

for u in range(3):
    for e in range(u+1,4):
        logreg = LogisticRegression(solver="lbfgs",multi_class='multinomial',penalty='l2',max_iter=500, random_state=3)
        a = data_test[:,u]
        b = data_test[:,e]
        c = np.c_[a,b]
        logreg.fit(c, labels_test)
        x_min, x_max = X[:, u].min() - 0.5, X[:, u].max() + 0.5
        y_min, y_max = X[:, e].min() - 0.5, X[:, e].max() + 0.5
        h = 0.02
        xx, yy = np.meshgrid(np.arange(x_min, x_max, h), np.arange(y_min, y_max, h)) #meshgrid to divide the plane in many pieces to compute the higher probability
        Z = logreg.predict(np.c_[xx.ravel(), yy.ravel()])

        
        Z = Z.reshape(xx.shape)
        fig, ax = plt.subplots(figsize=(10,8));
        plt.pcolormesh(xx, yy, Z, cmap=plt.cm.Set3,shading='auto')

        
        scatter = ax.scatter(X[:, u], X[:, e], c=t, edgecolors="k", cmap=plt.cm.Set3);
        
        legend1 = ax.legend(*scatter.legend_elements(), loc="best", fontsize=18, title='Classes');
        ax.add_artist(legend1);
        
        plt.xlabel(feature_names[u])
        plt.ylabel(feature_names[e])
        
        plt.xlim(xx.min(), xx.max())
        plt.ylim(yy.min(), yy.max())
        plt.title(feature_names[e]+' vs '+feature_names[u]);

        fig.savefig(f"boundary_decision_surfaces_{e}.png", dpi=300, bbox_inches="tight")
        plt.show();
````
